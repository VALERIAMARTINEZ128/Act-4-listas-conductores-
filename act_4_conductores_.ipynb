```python
# ==============================================================
#           LISTA INICIAL DE CONDUCTORES (MÍNIMO 4)
# ==============================================================

# Se inicia con una lista que contiene los diccionarios de los conductores
conductores = [
    {
        "Nombre": "Carlos Gomez",
        "Licencia": "Federal Tipo A",
        "Ruta": "CDMX - Guadalajara",
        "Placas": "72-RA-5X",
        "Horas": "45",
        "Empresa": "Transportes Express",
        "Seguridad": "Aprobado"
    },
    {
        "Nombre": "Ana Martinez",
        "Licencia": "Estatal Tipo B",
        "Ruta": "Monterrey - Laredo",
        "Placas": "11-BB-8Z",
        "Horas": "38",
        "Empresa": "Logistica del Norte",
        "Seguridad": "Aprobado"
    },
    {
        "Nombre": "Luis Rodriguez",
        "Licencia": "Federal Tipo A",
        "Ruta": "Veracruz - Puebla",
        "Placas": "45-MK-2Y",
        "Horas": "50",
        "Empresa": "Carga Segura",
        "Seguridad": "Pendiente"
    },
    {
        "Nombre": "Sofia Hernandez",
        "Licencia": "Estatal Tipo B",
        "Ruta": "Mérida - Cancún",
        "Placas": "88-TT-3W",
        "Horas": "30",
        "Empresa": "Ruta Maya S.A.",
        "Seguridad": "Aprobado"
    }
]

# ==============================================================
#                        FUNCIONES CRUD
# ==============================================================

def agregar_conductor(lista):
    print("\n--- 1. AGREGAR CONDUCTOR ---")

    # Se crea un diccionario vacío para rellenar los datos del nuevo conductor
    diccionario_conductor = {}

    # Captura de datos interactiva por medio de inputs
    diccionario_conductor["Nombre"] = input("Nombre del conductor: ").strip()
    diccionario_conductor["Licencia"] = input("Tipo de licencia: ").strip()
    diccionario_conductor["Ruta"] = input("Ruta asignada: ").strip()
    diccionario_conductor["Placas"] = input("Placas del vehículo: ").strip()
    diccionario_conductor["Horas"] = input("Horas de manejo: ").strip()
    diccionario_conductor["Empresa"] = input("Empresa de transporte: ").strip()
    diccionario_conductor["Seguridad"] = input("Registro de seguridad: ").strip()

    # Se agrega el nuevo diccionario a la lista principal
    lista.append(diccionario_conductor)
    print(f"\n¡Conductor {diccionario_conductor['Nombre']} agregado con éxito!")

    # Se muestra la tabla actualizada automáticamente (GigaNota / MegaNota)
    listar_conductor(lista)


def listar_conductor(lista):
    print("\n=========================================================================================================================")
    print(f"{'No.':<4} | {'NOMBRE':<18} | {'LICENCIA':<15} | {'RUTA ASIGNADA':<20} | {'PLACAS':<10} | {'HORAS':<6} | {'EMPRESA':<20} | {'SEGURIDAD':<10}")
    print("=========================================================================================================================")

    if not lista:
        print("   No hay conductores registrados en el sistema.")
        print("=========================================================================================================================")
        return

    # Recorremos la lista de diccionarios
    for indice, conductor in enumerate(lista, 1):
        # Imprimimos el índice alineado
        print(f"{indice:<4} | ", end="")

        # Desempaquetamos usando FOR KEY, VALUE (Clave-Valor) como se solicitó
        for key, value in conductor.items():
            # Formateamos el espacio de cada columna dependiendo del campo
            if key == "Nombre":
                print(f"{value:<18} | ", end="")
            elif key == "Licencia":
                print(f"{value:<15} | ", end="")
            elif key == "Ruta":
                print(f"{value:<20} | ", end="")
            elif key == "Placas":
                print(f"{value:<10} | ", end="")
            elif key == "Horas":
                print(f"{value:<6} | ", end="")
            elif key == "Empresa":
                print(f"{value:<20} | ", end="")
            elif key == "Seguridad":
                print(f"{value:<10}", end="")
        print() # Salto de línea por cada fila de la tabla
    print("=========================================================================================================================\n")


def buscar_conductor(lista):
    print("\n--- 3. BUSCAR CONDUCTOR ---")
    nombre_buscar = input("Escribe el nombre del conductor a buscar: ").strip().lower()

    conductor_encontrado = []

    for conductor in lista:
        if conductor["Nombre"].lower() == nombre_buscar:
            conductor_encontrado.append(conductor)

    if conductor_encontrado:
        print("\n¡Resultados de la búsqueda!")
        listar_conductor(conductor_encontrado)
    else:
        print(f"\nNo se encontró ningún conductor con el nombre: {nombre_buscar}")


def actualizar_conductor(lista):
    print("\n--- 5. ACTUALIZAR CONDUCTOR ---")
    nombre_buscar = input("Escribe el nombre del conductor a modificar: ").strip().lower()

    encontrado = False
    for conductor in lista:
        if conductor["Nombre"].lower() == nombre_buscar:
            encontrado = True
            print(f"\nModificando a: {conductor['Nombre']}. (Presiona ENTER para mantener el valor actual)")

            nuevo_nombre = input(f"Nuevo Nombre [{conductor['Nombre']}]: ").strip()
            nueva_licencia = input(f"Nuevo Tipo de Licencia [{conductor['Licencia']}]: ").strip()
            nueva_ruta = input(f"Nueva Ruta [{conductor['Ruta']}]: ").strip()
            nuevas_placas = input(f"Nuevas Placas [{conductor['Placas']}]: ").strip()
            nuevas_horas = input(f"Nuevas Horas de Manejo [{conductor['Horas']}]: ").strip()
            nueva_empresa = input(f"Nueva Empresa [{conductor['Empresa']}]: ").strip()
            nueva_seguridad = input(f"Nuevo Registro de Seguridad [{conductor['Seguridad']}]: ").strip()

            # Si el usuario escribió algo, se actualiza el campo, si dio ENTER se queda igual
            if nuevo_nombre: conductor["Nombre"] = nuevo_nombre
            if nueva_licencia: conductor["Licencia"] = nueva_licencia
            if nueva_ruta: conductor["Ruta"] = nueva_ruta
            if nuevas_placas: conductor["Placas"] = nuevas_placas
            if nuevas_horas: conductor["Horas"] = nuevas_horas
            if nueva_empresa: conductor["Empresa"] = nueva_empresa
            if nueva_seguridad: conductor["Seguridad"] = nueva_seguridad

            print("\n¡Datos actualizados con éxito!")
            break

    if not encontrado:
        print(f"\nNo se encontró al conductor: {nombre_buscar}")

    # Se muestra la tabla para verificar los cambios
    listar_conductor(lista)


def borrar_conductor(lista):
    print("\n--- 4. BORRAR CONDUCTOR ---")
    nombre_eliminar = input("Escribe el nombre del conductor a eliminar: ").strip().lower()

    encontrado = False
    for conductor in lista:
        if conductor["Nombre"].lower() == nombre_eliminar:
            lista.remove(conductor)
            print(f"\n¡El conductor {conductor['Nombre']} ha sido eliminado del registro!")
            encontrado = True
            break

    if not encontrado:
        print(f"\nNo se encontró al conductor: {nombre_eliminar}")

    # Se muestra la tabla para verificar la eliminación
    listar_conductor(lista)

# ==============================================================
#                        MENÚ PRINCIPAL
# ==============================================================

def menu():
    while True:
        print("=========================================")
        print("     SISTEMA DE CONTROL DE CONDUCTORES   ")
        print("=========================================")
        print("1. Agregar conductor en un diccionario")
        print("2. Listar diccionarios de conductor")
        print("3. Buscar conductor en diccionario")
        print("4. Borrar conductor en diccionario")
        print("5. Actualizar conductor en diccionario")
        print("6. Salir")
        print("=========================================")

        opcion = input("Selecciona una opción (1-6): ").strip()

        if opcion == "1":
            agregar_conductor(conductores)
        elif opcion == "2":
            listar_conductor(conductores)
        elif opcion == "3":
            buscar_conductor(conductores)
        elif opcion == "4":
            borrar_conductor(conductores)
        elif opcion == "5":
            actualizar_conductor(conductores)
        elif opcion == "6":
            print("\nSaliendo del programa... ¡Buen viaje!")
            break
        else:
            print("\nOpción no válida. Intenta de nuevo con un número del 1 al 6.\n")

# Ejecución del programa interactivo
if __name__ == "__main__":
    menu()
```
